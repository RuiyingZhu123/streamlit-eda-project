```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/amazon_sales_2025_INR.csv',
                 parse_dates=['Date'],
                 na_values=["NA", "", "null"])
```

```python
# Convert categorical fields to category dtype
cat_cols = [
    "Product_Category",
    "Product_Name",
    "Payment_Method",
    "Delivery_Status",
    "State",
    "Country"
]

for col in cat_cols:
    df[col] = df[col].astype("category")

df.dtypes
```

**Decision for duplicates**

No duplicates were found in previous phases. No rows need to be removed. 

**Decision for outliers**
  
Outliers are likely high-value purchases, so we keep them. 

### Hypothesis Generation for Cleaning Decisions

**1. Could type conversions introduce bias?**  
No — converting category-like fields to `category` does not modify the meaning of the data.

**2. Are outliers legitimate or errors?**  
Given the domain (Amazon Diwali season), large purchases and high-value electronics are normal.  
Outliers appear real transactions, not errors.

**3. What sensitivity analyses should be considered?**  
- Compare model performance with and without ouliered sales values.  

**4. What assumptions am I making?**
- Order_ID uniquely identifies each transaction.  
- No hidden missing values exist beyond what was detected in Phase 2.  
- High-value purchases are allowed.


### Iteration Signals for Future Phases

- If statistical EDA shows outliers strongly distort distributions → reconsider outlier treatment.
- If pricing patterns look inconsistent → return to Phase 2 and reassess type validity.
- If analyzing outcomes drastically change when excluding outliers → sensitivity analysis is needed.

Currently, no cleaning steps introduce major risks, so the cleaned dataset is ready for Phase 4.
